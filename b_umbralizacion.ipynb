**UMBRALIZAR IMAGENES**\
SE REFIERE A QUE CUANDO SE LLEGUE A UN VALOR DETERMINADO DEL PIXEL SE MUESTRE Y DE LO CONTRARIO NO LO HAGA O VICEVERSA

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
imagen_yugi = cv2.imread('img/yugi_collection.jpg')
imagen_yugi_GRAY = cv2.cvtColor(imagen_yugi,cv2.COLOR_BGR2GRAY)
imagen_yugi_RGB = cv2.cvtColor(imagen_yugi,cv2.COLOR_BGR2RGB)
plt.imshow(imagen_yugi_GRAY,cmap='gray')
```

```python
imagen_gray = np.ones(imagen_yugi_GRAY.shape,dtype='uint8')*50
```

**AUMENTAR Y DISMINUIR EL BRILLO DE LA IMAGEN**

```python
imagen_clara = cv2.add(imagen_gray,imagen_yugi_GRAY)
imagen_obscura = cv2.subtract(imagen_yugi_GRAY,imagen_gray)
```

**UMBRALIZACION THRESHOLD**\
**(_,) se utiliza aqu para ignorar el valor del umbral devuelto por cv2.threshold, ya que solo estamos interesados en la imagen umbralizada.**

```python
#imagen clara
_,img_thresh_clara = cv2.threshold(imagen_clara,190,255,cv2.THRESH_BINARY)
_,img_thresh_clara2 = cv2.threshold(imagen_clara,190,255,cv2.THRESH_BINARY_INV)
img_adaptativa_clara = cv2.adaptiveThreshold(imagen_clara,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,7)

#imagen obscura
_,img_thresh_obscura = cv2.threshold(imagen_obscura,50,255,cv2.THRESH_BINARY)
_,img_thresh_obscura2 = cv2.threshold(imagen_obscura,50,255,cv2.THRESH_BINARY_INV)
img_adaptativa_obscura = cv2.adaptiveThreshold(imagen_obscura,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,11,7)
```

```python
imagen_comparativa = plt.figure(figsize=(25,20))

#clara
x1 = imagen_comparativa.add_subplot(1,4,1)
x1.imshow(imagen_clara,cmap='gray')
x1.set_title("original clara") 


x2 = imagen_comparativa.add_subplot(1,4,2)
x2.imshow(img_thresh_clara,cmap='gray')
x2.set_title("clara threshold 1")


x3 = imagen_comparativa.add_subplot(1,4,3)
x3.imshow(img_thresh_clara2,cmap='gray')
x3.set_title("clara threshold 2")


x4 = imagen_comparativa.add_subplot(1,4,4)
x4.imshow(img_adaptativa_clara,cmap='gray')
x4.set_title("clara threshold comparativa")

#obscura
x5 = imagen_comparativa.add_subplot(2,4,1)
x5.imshow(imagen_obscura,cmap='gray')
x5.set_title("original obscura") 


x6 = imagen_comparativa.add_subplot(2,4,2)
x6.imshow(img_thresh_obscura,cmap='gray')
x6.set_title("obscura threshold 1")


x7 = imagen_comparativa.add_subplot(2,4,3)
x7.imshow(img_thresh_obscura2,cmap='gray')
x7.set_title("obscura threshold 2")


x8 = imagen_comparativa.add_subplot(2,4,4)
x8.imshow(img_adaptativa_obscura,cmap='gray')
x8.set_title("obscura threshold comparativa")
plt.show()

```
