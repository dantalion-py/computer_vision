TRANSFORMACION Y ANOTACION DE IMAGENES

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import seaborn as sns
```

RECORTE, TOMAR UN FRAGMENTO DE PIXELES DE LA IMAGEN Y DESECHAR LOS DEMAS

```python
img = 100*np.ones((10,10,3),np.uint8)
img2 = 100*np.ones((10,10,3),np.uint8)
```

FIGURA

```python
#MODIFICACION
img[4,4,0] = 255
img[5,4,0] = 80
img[5,5,0] = 70
img[4,5,0] = 140

img[4,4,1] = 60
img[5,4,1] = 33
img[5,5,1] = 70
img[4,5,1] = 10

img[4,4,2] = 255
img[5,4,2] = 90
img[5,5,2] = 170
img[4,5,2] = 240

#RECORTE
recorte = img[3:7,3:7]

#IMAGEN
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.set_title("IMAGEN")
ax1.imshow(img)
#IMAGEN RECORTE
ax2 = fig.add_subplot(1,2,2)
ax2.set_title("RECORTE")
ax2.imshow(recorte)
plt.show()
```

RECORTE IMAGEN 

```python
img_tastu = cv2.imread('img/tatsumaki.png',1)
imgRGB = cv2.cvtColor(img_tastu,cv2.COLOR_BGR2RGB)
#RECORTE primer argumento rows y segundo columns
recorte_tatsumaki = imgRGB[255:630,450:710]
#IMAGEN
fig2 = plt.figure()
ax_2 = fig2.add_subplot(1,2,1)
ax_2.set_title("IMAGEN")
ax_2.imshow(imgRGB)
#IMAGEN RECORTE
ax_2 = fig2.add_subplot(1,2,2)
ax_2.set_title("RECORTE")
ax_2.imshow(recorte_tatsumaki)
plt.show()
```

```python
#INTERPOLACION DE IMAGENES PUEDE HACER MAS GRANDE O MAS PEQUENA UNA IMAGEN asi como mas ancha o mas corta
alto = 500
ancho = 1500
dimension = (alto, ancho)
imagen  = cv2.imread('img/kobayashi.jpg')
imagen_kobayashi = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
inter_1 = cv2.resize(imagen_kobayashi,None,fx=0.3,fy=0.3)
inter_2 = cv2.resize(imagen_kobayashi,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
inter_3 = cv2.resize(imagen_kobayashi,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_CUBIC)
#MODIFICACION
inter_MOD = cv2.resize(imagen_kobayashi,dimension,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)

#original
# sns.set()
compare_img = plt.figure(figsize=(15,7))
x1 = compare_img.add_subplot(1,4,1)
x1.set_title("img original")
x1.imshow(imagen_kobayashi)


# #img1
x2 = compare_img.add_subplot(1,4,2)
x2.set_title("img 1")
x2.imshow(inter_1)


# #img2
x3 = compare_img.add_subplot(1,4,3)
x3.set_title("img 2")
x3.imshow(inter_2)

# #img3
x4 = compare_img.add_subplot(1,4,4)
x4.set_title("img 3")
x4.imshow(inter_3)


# cv2.imshow('BGR',inter_1)
cv2.imshow('RGB',inter_MOD)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
