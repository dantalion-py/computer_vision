```python
import cv2
```

```python
video_record = cv2.VideoCapture(0)
while True:
    ret,frame = video_record.read()
    hvs = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    edg = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

    print(ret)

    cv2.imshow('VIDEO CAPTURA RGB',frame)
    cv2.imshow('VIDEO CAPTURA HVS',hvs)
    cv2.imshow('VIDEO CAPTURA EDG',edg)
    t = cv2.waitKey(1)
    if 27 == t:
        break


video_record.release()
cv2.destroyAllWindows()
```
