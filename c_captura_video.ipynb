```python
import cv2
```

**CAPTURA DE IMAGENES POR LA WEBCAM**

```python
#CREAR LA VIDEO CAPTURA
captura_video = cv2.VideoCapture(0)

#CICLO PARA EJECUTAR LOS FRAMES
while True:
    ret,frame = captura_video.read()
    print(ret)
    #MOSTRAR FRAMES
    cv2.imshow("CAPTURA DE VIDEO", frame)
    t = cv2.waitKey(1)
    #27 ES IGUAL A (ESC) 32 ES (SPACE)
    if t == 32:
        break
captura_video.release()
cv2.destroyAllWindows()
```
