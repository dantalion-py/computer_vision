```python
import matplotlib.pyplot as plt
import cv2
```

```python
# imgRGB = cv2.imread('img/tatsumaki.png')
img = cv2.imread('img/tatsumaki.png',1)
imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

```

EXTRAER CANALES DE COLOR opcion1


```python
R = imgRGB[:,:,0]
G = imgRGB[:,:,1]
B = imgRGB[:,:,2]
plt.imshow(imgRGB)
```

EXTRAER CANALES DE COLOR opcion 2

```python
R,G,B = cv2.split(imgRGB)
print(R)
```

MOSTRAR LAS IMAGENES\
TAMBIEN SE PUEDE CAMBIAR EL VALOR DEL CANAL Y GUARDAR LA IMAGEN

```python
fig = plt.figure()
#Canal rojo
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(R,cmap="gray")
ax1.set_title("CANAL ROJO")
#Canal verde
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(G,cmap="gray")
ax2.set_title("CANAL VERDE")
#Canal azul
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(B,cmap="gray")
ax3.set_title("CANAL AZUL")


#RECOSTRUCCION
imgre = cv2.merge((R,G+100,B))


#IMAGEN ORIGINAL
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(imgre)
ax4.set_title("CANAL original")


#GUARDAR IMAGEN
cv2.imwrite('img/tatsumakiGREEN_MOD.png',imgre)
plt.show()
```

Conversion de RGB a HSV

```python
imgHSV = cv2.cvtColor(imgRGB,cv2.COLOR_RGB2HSV)
```

```python
H,S,V = cv2.split(imgHSV)
fig = plt.figure()
#Canal rojo
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(H,cmap="gray")
ax1.set_title("CANAL H")
#Canal verde
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(S,cmap="gray")
ax2.set_title("CANAL S")
#Canal azul
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(V,cmap="gray")
ax3.set_title("CANAL V")


#RECOSTRUCCION
imgre = cv2.merge((H,S,V))


#IMAGEN ORIGINAL
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(imgre)
ax4.set_title("CANAL original")
```
