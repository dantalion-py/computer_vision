OPERACIONES LOGICAS

```python
import matplotlib.pyplot as plt
import cv2
```

```python
circle_img = cv2.imread('img/circle_a.jpg',0)
triangle_img = cv2.imread('img/triangle.png',0)

width = 500
height = 500
size = (width,height)
new_image_circle = cv2.resize(circle_img,size)
new_image_triangle = cv2.resize(triangle_img,size)
cv2.imwrite('img/circle_mod.jpg',new_image_circle)
cv2.imwrite('img/triangle_mod.jpg',new_image_triangle)
plt.imshow(triangle_img)
```

```python
img_1 = cv2.bitwise_and(new_image_triangle,new_image_circle,mask=None)
img_2 = cv2.bitwise_or(new_image_triangle,new_image_circle,mask=None)
img_3 = cv2.bitwise_xor(new_image_triangle,new_image_circle,mask=None)
figure_operations = plt.figure(figsize=(10,30))
x1 = figure_operations.add_subplot(1,5,1)
x1.set_title("circle")
x1.imshow(new_image_circle,cmap='gray')

x2 = figure_operations.add_subplot(1,5,2)
x2.set_title("triangle")
x2.imshow(new_image_triangle,cmap='gray')

x3 = figure_operations.add_subplot(1,5,3)
x3.set_title("and")
x3.imshow(img_1,cmap='gray')

x4 = figure_operations.add_subplot(1,5,4)
x4.set_title("or")
x4.imshow(img_2,cmap='gray')

x5 = figure_operations.add_subplot(1,5,5)
x5.set_title("xor")
x5.imshow(img_3,cmap='gray')

plt.show()
```

```python
img_pepsi = cv2.imread('img/pepsi.png',1)
img_fond = cv2.imread('img/fond.jpg',1)
height_mod = 500
width_mod = 500
size_mod = (width_mod,height_mod)


img_fond_mod = cv2.resize(img_fond,size_mod)
img_pepsi_mod = cv2.resize(img_pepsi,size_mod)
#CORRECCION DE COLOR
img_fond = cv2.cvtColor(img_fond_mod,cv2.COLOR_BGR2RGB)
img_pepsi = cv2.cvtColor(img_pepsi_mod,cv2.COLOR_BGR2RGB)


# cv2.imwrite('img/logo_mod.jpg',img_fond_mod)
# cv2.imwrite('img/pepsi_mod.jpg',img_pepsi_mod)
```

```python
img_and_pepsi = cv2.bitwise_or(img_fond,img_pepsi,mask=None)
plt.imshow(img_and_pepsi)
```

**UMBRALIZACION**\
**CREACION DE UNA MASCARA**


```python
img_logo_mask_gray = cv2.cvtColor(img_fond,cv2.COLOR_RGB2GRAY)
_,img_mask = cv2.threshold(img_logo_mask_gray,150,255,cv2.THRESH_BINARY)
plt.imshow(img_mask)
```

**MASCARA INVERTIDA**

```python
img_reverse_mask = cv2.bitwise_not(img_mask)
plt.imshow(img_reverse_mask)
```

**OPERACIONES DE IMAGENES**

```python
img_gap_p1 = cv2.bitwise_and(img_pepsi,img_fond,mask=img_mask)
img_gap_p2 = cv2.bitwise_and(img_pepsi,img_fond,mask=img_reverse_mask)
img_gap_p3 = cv2.subtract(img_fond_mod,img_gap_p2)
# img_gap_p3 = cv2.add(img_fond_mod,img_gap_p2)
plt.imshow(img_gap_p3)
```

**IMAGEN COMPARATIVA**

```python
img_collection = plt.figure(figsize=(15,20))
x1 = img_collection.add_subplot(1,7,1)
x1.imshow(img_fond)
x1.set_title('original fond')

x2 = img_collection.add_subplot(1,7,2)
x2.imshow(img_pepsi)
x2.set_title('original pepsi')

x3 = img_collection.add_subplot(1,7,3)
x3.imshow(img_mask)
x3.set_title('mascara')

x4 = img_collection.add_subplot(1,7,4)
x4.imshow(img_reverse_mask)
x4.set_title('mascara reverse')

x5 = img_collection.add_subplot(1,7,5)
x5.imshow(img_gap_p1)
x5.set_title('and 1')

x6 = img_collection.add_subplot(1,7,6)
x6.imshow(img_gap_p2)
x6.set_title('and 2')

x7 = img_collection.add_subplot(1,7,7)
x7.imshow(img_gap_p3)
x7.set_title('add')
    
plt.show()
```
