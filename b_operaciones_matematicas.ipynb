```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#IMAGEN CARTS
yugi_carts = cv2.imread('img/yugi_collection.jpg',1)
yugi_carts_RGB = cv2.cvtColor(yugi_carts,cv2.COLOR_BGR2RGB)
yugi_carts_GRAY = cv2.cvtColor(yugi_carts,cv2.COLOR_BGR2GRAY)
plt.imshow(yugi_carts_RGB)
```

CREAR UNA MATRIX EN ESCALA DE GRISES DEL TAMAO DE LA IMAGEN

```python
matriz_GRY = np.ones(yugi_carts_GRAY.shape,dtype='uint8')*100
matriz_RGB = np.ones(yugi_carts_RGB.shape,dtype='uint8')*100
plt.imshow(matriz_RGB)
```

AUMENTAR EL BRILLO DE LA IMAGEN RGB

```python
brillo_rbg = cv2.add(yugi_carts_RGB,matriz_RGB)
brillo_rbg_yugi = cv2.cvtColor(brillo_rbg,cv2.COLOR_BGR2RGB)
plt.imshow(brillo_rbg)
```

DISMINUIR EL BRILLO DE LA IMAGEN RGB

```python
disbrillo_rbg = cv2.subtract(yugi_carts_RGB,matriz_RGB)
disbrillo_rbg_yugi = cv2.cvtColor(disbrillo_rbg,cv2.COLOR_BGR2RGB)
plt.imshow(disbrillo_rbg)
```

AUMENTAR BRILLO GRAY

```python
brillo_gray = cv2.add(yugi_carts_GRAY,matriz_GRY)
plt.imshow(brillo_gray,cmap='gray')
```

DISMINUIR BRILLO GRAY

```python
disbrillo_gray = cv2.subtract(yugi_carts_GRAY,matriz_GRY)
plt.imshow(brillo_gray,cmap='gray')
```

```python
all_contrast = plt.figure(figsize=(20,20))
x1 = all_contrast.add_subplot(1,3,1)
x1.imshow(yugi_carts_RGB)
x1.set_title('ORIGINAL')

#CLARO rgb
x2 = all_contrast.add_subplot(1,3,2)
x2.imshow(brillo_rbg)
x2.set_title('CLARO')

#OBSCURO rgb
x3 = all_contrast.add_subplot(1,3,3)
x3.imshow(disbrillo_rbg)
x3.set_title('OBSCURO')

#ORIGINAL gray
x4 = all_contrast.add_subplot(2,3,1)
x4.imshow(yugi_carts_GRAY, cmap='gray')
x4.set_title('ORIGINAL')

#CLARO gray
x5 = all_contrast.add_subplot(2,3,2)
x5.imshow(brillo_gray, cmap='gray')
x5.set_title('CLARO')

#OBSCURO gray
x6 = all_contrast.add_subplot(2,3,3)
x6.imshow(disbrillo_gray, cmap='gray')
x6.set_title('OBSCURO')
```
